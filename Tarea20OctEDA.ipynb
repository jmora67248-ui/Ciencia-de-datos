#Ingenieria de Caracteristicas
Esl proceso de usar el dominio del conocimiento para crear nuevas caracteristicas (feature vector) que logre que los algoritmos de machine learning funcionen adecuadamente.

```python
def extract_title(data):
  titles=data["Name"].str.extract("([A-Za-z]+)\.",expand=True)
  return titles
```

```python
df["Title"]=extract_title(df)
```

```python
df["Title"].value_counts()
```

### creando un mapeo de titulos

Para ello reemplazamos titulos de la siguiente manera

* Mr=0
* Miss=1
* Mrs=2
* Otros=3

```python
def replace_titles(data):
  title_mapping={"Mr":0,"Miss":1, "Mrs":2}
  data["Title"]=data["Title"].map(title_mapping).fillna(3)
  return data
```

```python
replace_titles(df)
```

```python
df[["Name","Title"]].head()
```

```python
# Explorar estadsticas bsicas de Fare
print(df['Fare'].describe())
print("\nValores nulos en Fare:", df['Fare'].isnull().sum())

# Visualizar distribucin de Fare
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
sns.histplot(data=df, x='Fare', bins=50, kde=True)
plt.title('Distribucin de Fare')

plt.subplot(1, 2, 2)
sns.boxplot(data=df, y='Fare')
plt.title('Boxplot de Fare')
plt.tight_layout()
plt.show()

# Anlisis de Fare por supervivencia
facet = sns.FacetGrid(df, hue="Survived", aspect=4)
facet.map(sns.kdeplot, "Fare", shade=True)
facet.set(xlim=(0, df['Fare'].max()))
facet.add_legend()
plt.show()

# Estadsticas por grupo de supervivencia
print("\nFare promedio por supervivencia:")
print(df.groupby('Survived')['Fare'].mean())
```

* # Existe una clara correlacin positiva entre la tarifa y la supervivencia. El precio promedio pagado por los supervivientes (48.39) fue ms del doble que el de los no supervivientes (22.11). La distribucin de los supervivientes (naranja) es ms plana y se extiende mucho ms hacia las tarifas altas.

#Binning Cuartiles

```python
# Opcin 1: Binning por cuartiles
df['Fare_category'] = pd.qcut(df['Fare'], q=4, labels=['Bajo', 'Medio', 'Alto', 'Muy Alto'])

# Opcin 2: Binning manual (basado en la distribucin del Titanic)
# bins = [0, 7.91, 14.45, 31, float('inf')]
# df['Fare_category'] = pd.cut(df['Fare'], bins=bins, labels=['Bajo', 'Medio', 'Alto', 'Muy Alto'])

# Ver distribucin
print(df['Fare_category'].value_counts())

# Visualizar Fare_category vs Survived
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='Fare_category', hue='Survived')
plt.title('Supervivencia por Categora de Fare')
plt.show()

# Tasa de supervivencia por categora
fare_survival = df.groupby('Fare_category')['Survived'].mean()
print("\nTasa de supervivencia por categora de Fare:")
print(fare_survival)
```

#Anlisis de puerto de embarque

```python
# Explorar valores en Embarked
print("Valores en Embarked:")
print(df['Embarked'].value_counts())
print("\nValores nulos en Embarked:", df['Embarked'].isnull().sum())

# Rellenar valores nulos con la moda (ms frecuente)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Visualizar Embarked vs Survived
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='Embarked', hue='Survived')
plt.title('Supervivencia por Puerto de Embarque')
plt.show()

# Tasa de supervivencia por puerto
embarked_survival = df.groupby('Embarked')['Survived'].mean()
print("\nTasa de supervivencia por puerto:")
print(embarked_survival)
```

#A partir de la grfica, se concluye que el puerto de embarque ("Embarked") est fuertemente correlacionado con la supervivencia.

#Aunque la gran mayora de los pasajeros embarc en Southampton ('S'), este grupo tuvo la tasa de supervivencia ms baja (aproximadamente 34%). Por el contrario, los pasajeros que embarcaron en Cherbourg ('C') tuvieron la tasa de supervivencia ms alta, superando el 55%.

```python
# Crosstab de Embarked y Fare_category
print(pd.crosstab(df['Embarked'], df['Fare_category']))

# Fare promedio por puerto
print("\nFare promedio por puerto:")
print(df.groupby('Embarked')['Fare'].mean())

# Visualizacin combinada
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Embarked', y='Fare', hue='Survived', ci=None)
plt.title('Fare Promedio por Puerto y Supervivencia')
plt.show()

# Facet grid: Embarked vs Fare colored by Survived
g = sns.FacetGrid(df, col='Embarked', hue='Survived', height=4, aspect=1.2)
g.map(sns.kdeplot, 'Fare', shade=True)
g.add_legend()
g.set(xlim=(0, 100))
plt.show()
```

##El puerto de embarque es un fuerte indicador de la tarifa pagada. Los pasajeros que embarcaron en Cherbourg ('C') pagaron, en promedio, las tarifas ms altas (59.95), con una alta concentracin de pasajeros en las categoras "Alto" y "Muy Alto". Por el contrario, los pasajeros de Queenstown ('Q') pagaron las tarifas ms bajas (13.27), estando la mayora en la categora "Bajo".


##La relacin entre tarifa y supervivencia cambia segn el puerto. En Cherbourg ('C') y Southampton ('S'), los supervivientes pagaron tarifas significativamente ms altas que los no supervivientes.


##Sin embargo, en Queenstown ('Q'), la tarifa fue uniformemente baja para ambos grupos (supervivientes y no supervivientes), lo que sugiere que la tarifa no fue un factor determinante para la supervivencia en ese puerto especfico.

```python
# Feature: Fare por persona (considerando familia)
df['Family_Size'] = df['SibSp'] + df['Parch'] + 1
df['Fare_per_person'] = df['Fare'] / df['Family_Size']

# Feature: Viaja solo?
df['Is_Alone'] = (df['Family_Size'] == 1).astype(int)

# Feature: Combinacin de Embarked y Pclass
df['Embarked_Pclass'] = df['Embarked'].astype(str) + '_' + df['Pclass'].astype(str)

# Feature: Pag tarifa alta?
df['High_Fare'] = (df['Fare'] > df['Fare'].median()).astype(int)

# Mostrar las nuevas features
print(df[['Fare', 'Family_Size', 'Fare_per_person', 'Is_Alone', 'High_Fare']].head(10))
```

```python
# Matriz de correlacin con las nuevas features
features_to_analyze = ['Survived', 'Pclass', 'Age', 'SibSp', 'Parch',
                       'Fare', 'Family_Size', 'Fare_per_person', 'Is_Alone', 'High_Fare']

plt.figure(figsize=(12, 8))
sns.heatmap(df[features_to_analyze].corr(), annot=True, cmap='coolwarm', center=0)
plt.title('Matriz de Correlacin')
plt.show()
```

## Los coeficientes que mas tienen relacin con survived son: Fare(0.26),Highfare(0.27)

##Segn la matriz de correlacin, el coeficiente entre "Is_Alone" y "Survived" es -0.2.

##Esto indica una correlacin negativa dbil. En el contexto del anlisis, significa que las personas que viajaban solas ("Is_Alone" = 1) tuvieron una ligera tendencia menor a sobrevivir que aquellas que viajaban acompaadas ("Is_Alone" = 0).
